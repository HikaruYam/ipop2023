```python
import networkx as nx
import plotly.graph_objects as go
from IPython.display import display
import time

def create_initial_figure(G):
    edge_x = []
    edge_y = []
    for edge in G.edges():
        x0, y0 = G.nodes[edge[0]]['pos']
        x1, y1 = G.nodes[edge[1]]['pos']
        edge_x += [x0, x1, None]
        edge_y += [y0, y1, None]

    edge_trace = go.Scatter(x=edge_x, y=edge_y, line=dict(width=2, color='black'), hoverinfo='none', mode='lines')

    packet_trace = go.Scatter(x=[], y=[], mode='markers', hoverinfo='none', marker=dict(size=25, color='red'))

    node_x = []
    node_y = []
    for node in G.nodes():
        x, y = G.nodes[node]['pos']
        node_x.append(x)
        node_y.append(y)

    node_trace = go.Scatter(x=node_x, y=node_y, mode='markers+text', hoverinfo='text', marker=dict(size=50, color='skyblue', symbol='square'), text=list(G.nodes), textposition='top center')

    layout = go.Layout(showlegend=False,
                       hovermode='closest',
                       margin=dict(t=0, b=0, l=0, r=0),
                       xaxis=dict(showgrid=False, zeroline=False, showticklabels=False, color='white'),
                       yaxis=dict(showgrid=False, zeroline=False, showticklabels=False, color='white'))

    fig = go.FigureWidget(data=[edge_trace, packet_trace, node_trace], layout=layout)
    
    return fig

def update_fig(fig, G, packet_transmission=False):
    edge_x = []
    edge_y = []
    for edge in G.edges():
        x0, y0 = G.nodes[edge[0]]['pos']
        x1, y1 = G.nodes[edge[1]]['pos']
        edge_x += [x0, x1, None]
        edge_y += [y0, y1, None]

    if packet_transmission:
        packet_x = [(G.nodes[1]['pos'][0] + G.nodes[2]['pos'][0]) / 2]
        packet_y = [(G.nodes[1]['pos'][1] + G.nodes[2]['pos'][1]) / 2]
    else:
        packet_x = []
        packet_y = []

    node_x = []
    node_y = []
    for node in G.nodes():
        x, y = G.nodes[node]['pos']
        node_x.append(x)
        node_y.append(y)

    fig.data[0]['x'] = edge_x
    fig.data[0]['y'] = edge_y
    fig.data[1]['x'] = packet_x
    fig.data[1]['y'] = packet_y
    fig.data[2]['x'] = node_x
    fig.data[2]['y'] = node_y

def main():
    G = nx.DiGraph()
    G.add_nodes_from([0, 1, 2, 3, 4, 5, 6])
    G.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 5), (1, 6), (0,1)])
    pos = {
        0: (1, 0),
        1: (0, 0),
        2: (-1, 2),
        3: (-1, 1),
        4: (-1, 0),
        5: (-1, -1),
        6: (-1, -2)
    } 
    for node, position in pos.items():
        G.nodes[node]['pos'] = position

    fig = create_initial_figure(G)
    display(fig)

    while True:
        update_fig(fig, G, packet_transmission=False)
        time.sleep(1)
        update_fig(fig, G, packet_transmission=True)
        time.sleep(1)

if __name__ == "__main__":
    main()
```
